```python
import numpy as np 
import pandas as pd 
```

```python
df1=pd.read_csv('accidents_2005_to_2007.csv')
df1
```

```python
df2=pd.read_csv('accidents_2009_to_2011.csv')
df2
```

```python
df3=pd.read_csv('accidents_2012_to_2014.csv')
df3
```

```python
df_all = pd.concat([df1, df2, df3], ignore_index=True)
print(df1.shape[0]+df2.shape[0]+df3.shape[0])
```

```python
df_all
```

```python
import matplotlib.pyplot as plt
plt.scatter(df_all['Longitude'], df_all['Latitude'])
plt.figure(figsize=(30, 30))
```

```python
counts = df_all['Weather_Conditions'].value_counts()
plt.figure(figsize=(15, 8))
plt.bar(counts.index, counts.values)
plt.xticks(rotation=45)
```
